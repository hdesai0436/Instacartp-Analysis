## create for loop define busiest data 

```python
result = []
for name in df_merge_ord_pro['orders_day_of_the_week']:
    if name == 0:
        result.append("Busiest day")
    elif name == 4:
        result.append("Slowest Day")
    else:
        result.append("regularbusy day")

    
```

```python
df_merge_ord_pro['busiest_day'] = result
```

```python
df_merge_ord_pro.head()
```

## create for loop to define mutiples busiest day

```python
result_2 = []
for val in df_merge_ord_pro['orders_day_of_the_week']:
    if val == 0 or val == 1:
        result_2.append('Busiest day')
    elif val == 3 or val == 4:
        result_2.append('slowest day')
    else:
        result_2.append('regular day')
        
        
```

```python
df_merge_ord_pro['busiest_days'] = result_2
```

```python
df_merge_ord_pro.head()
```

```python
df_merge_ord_pro['busiest_days'].value_counts()
```

```python
df_merge_ord_pro['order_hour_of_day'].value_counts()
```

## create function that fine most orders in the hours

```python

result_3 = []
for name in df_merge_ord_pro['order_hour_of_day']:
    if name in [10,11,14,15,13,12,16,9]:
        result_3.append('most orders')
    elif name in [3,4,2,5,1]:
        result_3.append('Fewest orders')
    else:
        result_3.append('Average orders')
        
        
```

```python
df_merge_ord_pro['busiest_period_of_day'] = result_3
```

```python
df_merge_ord_pro['busiest_period_of_day'].value_counts()
```

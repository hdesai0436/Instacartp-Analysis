## step 3

```python
df_merge['max_number_order'] = df_merge.groupby('user_id')['order_number'].transform(np.max)
```

```python
df_merge.head()
```

```python
df_merge['max_number_order'].value_counts()
```

```python
df_merge.loc[df_merge['max_number_order'] > 40, 'loyalty_flag'] = 'loyal_customer'
df_merge.loc[(df_merge['max_number_order'] <= 40) & (df_merge['max_number_order'] > 10), 'loyalty_flag'] = 'regular_customer'
df_merge.loc[df_merge['max_number_order'] <= 10, 'loyalty_flag'] = 'new_customer'
```

```python
df_merge['loyalty_flag'].value_counts(dropna=False)
```

## step 5

```python
df_merge.groupby(['loyalty_flag']).agg({'prices':['min','max','mean']})
```

## new customers spend on average 3 dollars from loyal_customer and $1 more from regular customer

## step 6

```python
df = df_merge.groupby(['user_id'])['prices'].mean().reset_index()
```

```python
df.head()
```

```python
df['mean'] = np.where(df['prices'] >= 10 , 'high_spender', 'low_spender')
```

```python
df1 = df_merge.merge(df[['user_id', 'mean']], on='user_id', how='left')
```

```python
df1['mean'].value_counts()
```

```python
df1.head()
```

```python
df_merge['mean_price_spent_customer'] = df_merge.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
df_merge[['user_id','mean_price_spent_customer']].head(20)
```

```python
df_merge.head(10)
```

```python
df_merge.loc[df_merge['mean_price_spent_customer'] >= 10, "spending_flag"] = 'High spender'
```

```python
df_merge.loc[df_merge['mean_price_spent_customer'] < 10, "spending_flag"] = 'Lower spender'
```

```python
df_merge['spending_flag'].value_counts()
```

```python
df_merge.head(20)
```

```python
df_merge.columns
```

```python
df_merge['customer_frequency'] =  df_merge.groupby('user_id')['days_since_last_order'].transform('median')
```

```python
df_merge[['user_id','customer_frequency']].head(10)
```

```python
df_merge.loc[df_merge['customer_frequency'] > 20 ,'frequency_flag'] = 'Non-frequent customer'
```

```python
df_merge.loc[(df_merge['customer_frequency'] > 10) & (df_merge['customer_frequency'] <= 20) ,'frequency_flag'] = 'Regular customer'
```

```python
df_merge.loc[df_merge['customer_frequency'] <= 10 ,'frequency_flag'] = 'Frequent customer'
```

```python
df_merge['frequency_flag'].value_counts(dropna=False)
```

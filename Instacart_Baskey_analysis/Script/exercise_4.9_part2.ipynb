## step 3

```python
#create histogram for order_hour_day
hist = df['order_hour_of_day'].plot.hist(bins=24)
```

## step 4

```python
#create bar chart for loyalty_flag to check count of loyal customers
bar = df['loyalty_flag'].value_counts().plot.bar()
```

## step 5

```python
#create sample dataframe 
sample_df = df.sample(n=100000,random_state=1)
```

```python
#group by hour of the day and calutate mean price
hour_price = sample_df.groupby('order_hour_of_day')['prices'].mean().reset_index()
```

```python
#create line chart
line_price_hour = sns.lineplot(data=sample_df,x='order_hour_of_day',y = 'prices')
```

## step 6 

```python
#create line chart for age and number of dependants
liner_dependants_age = sns.lineplot(data= sample_df,x='Age', y = 'Number_dependants')
```

## there is no clear correlation between age and number_dependats

## step 7

```python
##create a scatterplot of age and income
sns.scatterplot(x='Age', y='income', data=sample_df)
```

## income is increase after age of 40
